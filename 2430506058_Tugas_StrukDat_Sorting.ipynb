```python
# Latihan/Tugas Sorting

import time
import random

class Produk:
    def __init__(self, nama, harga, rating):
        self.nama = nama
        self.harga = harga
        self.rating = rating

    def __repr__(self):
        return f"{self.nama} (Harga: Rp{self.harga:,}, Rating: {self.rating}/5)"

def generate_produk(jumlah, kategori):
    nama_produk = ["Laptop", "HP", "Mouse", "Keyboard", "Monitor"]
    harga_produk = {
        "Laptop": (5000000, 25000000),
        "HP": (1000000, 20000000),
        "Mouse": (50000, 2000000),
        "Keyboard": (100000, 5000000),
        "Monitor": (1000000, 10000000)
    }

    if kategori not in nama_produk:
        return []

    produk_list = []
    for i in range(jumlah):
        nama = f"{kategori} {i+1}"
        harga = random.randint(*harga_produk[kategori])
        rating = round(random.uniform(1, 5), 1)
        produk_list.append(Produk(nama, harga, rating))
    return produk_list

def bubble_sort(arr, key='harga', ascending=True):
    if not arr:
        return []
    n = len(arr)
    for i in range(n-1):
        for j in range(n-i-1):
            if (getattr(arr[j], key) > getattr(arr[j+1], key)) == ascending:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

def quick_sort(arr, key='harga', ascending=True):
    if not arr:
        return []
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if getattr(x, key) < getattr(pivot, key)]
    middle = [x for x in arr if getattr(x, key) == getattr(pivot, key)]
    right = [x for x in arr if getattr(x, key) > getattr(pivot, key)]
    if ascending:
        return left + middle + right
    else:
        return right + middle + left

def tampilkan_produk(produk_list, limit=10):
    print("\nDaftar Produk:")
    for i, produk in enumerate(produk_list[:limit], 1):
        print(f"{i}. {produk}")
    if len(produk_list) > limit:
        print(f"... dan {len(produk_list)-limit} produk lainnya")

def main():
    nama_produk = ["Laptop", "HP", "Mouse", "Keyboard", "Monitor"]

    while True:
        kategori = input("Produk yang ingin dicari (Laptop, HP, Mouse, Keyboard, Monitor): ")
        if kategori in nama_produk:
            break
        print("Kategori tidak ditemukan! Coba lagi.")

    while True:
        try:
            jumlah_produk = int(input("Masukkan jumlah produk (1-100): "))
            if 1 <= jumlah_produk <= 100:
                break
            print("Jumlah produk harus antara 1-100.")
        except ValueError:
            print("Input harus berupa angka.")

    produk_list = generate_produk(jumlah_produk, kategori)
    if not produk_list:
        print("Terjadi kesalahan dalam pembuatan produk.")
        return

    while True:
        print("\n1. Urutkan berdasarkan Harga (Termurah ke Termahal)")
        print("2. Urutkan berdasarkan Harga (Termahal ke Termurah)")
        print("3. Keluar")

        try:
            pilihan = int(input("Masukkan pilihan: "))
            if pilihan == 3:
                print("Terima kasih telah menggunakan program ini!")
                break

            if pilihan == 1:
                produk_sorted = bubble_sort(produk_list, 'harga', True)
            elif pilihan == 2:
                produk_sorted = quick_sort(produk_list, 'harga', False)
            else:
                print("Pilihan tidak valid, masukkan angka antara 1-3.")
                continue

            print("\nHasil Sorting Harga:")
            tampilkan_produk(produk_sorted)

            while True:
                print("\nIngin mengurutkan berdasarkan rating?")
                print("1. Ya (Tertinggi ke Terendah)")
                print("2. Ya (Terendah ke Tertinggi)")
                print("3. Tidak, kembali ke menu utama")

                try:
                    pilihan_rating = int(input("Masukkan pilihan: "))
                    if pilihan_rating == 1:
                        produk_sorted = quick_sort(produk_sorted, 'rating', False)
                    elif pilihan_rating == 2:
                        produk_sorted = bubble_sort(produk_sorted, 'rating', True)
                    elif pilihan_rating == 3:
                        break
                    else:
                        print("Pilihan tidak valid, silakan pilih kembali.")
                        continue

                    print("\nHasil Sorting Rating:")
                    tampilkan_produk(produk_sorted)
                except ValueError:
                    print("Input harus berupa angka.")
        except ValueError:
            print("Input harus berupa angka.")

if __name__ == "__main__":
    main()
```
