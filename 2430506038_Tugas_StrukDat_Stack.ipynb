```python
class Warehouse:
    def __init__(self, max_capacity):
        self.max_capacity = max_capacity
        self.items = []

    def add_item(self, item):
        if len(self.items) >= self.max_capacity:
            print("Gudang Penuh! Tidak bisa menambah barang.")
            return
        self.items.append(item)
        print(f"Menambahkan barang: {item}")

    def remove_item(self):
        if not self.items:
            print("Gudang kosong, tidak ada barang yang bisa diambil!")
            return

        # Tampilkan daftar barang yang ada di gudang
        self.show_items()

        # Minta pengguna memilih indeks barang yang akan diambil
        index = int(input("Masukkan nomor barang yang ingin diambil: ")) - 1

        # Validasi indeks
        if index < 0 or index >= len(self.items):
            print("Indeks tidak valid!")
            return

        # Ambil barang berdasarkan indeks
        item = self.items.pop(index)
        print(f"Mengambil barang: {item}")

    def show_items(self):
        if not self.items:
            print("Gudang kosong.")
            return
        print("Barang dalam gudang:")
        for i, item in enumerate(self.items):
            print(f"{i + 1}. {item}")

# Input dua digit terakhir NPM untuk kapasitas gudang
npm = input("Masukkan dua digit terakhir NPM (kapasitas maksimal gudang): ")
max_capacity = int(npm)  # Kapasitas gudang adalah dua digit terakhir NPM

# Pengecekan kapasitas
while True:
    kapasitas = int(input("Masukkan kapasitas barang: "))
    if kapasitas > max_capacity:
        print("Gudang Penuh! Kapasitas melebihi batas. Silakan input ulang.")
    else:
        break  # Keluar dari loop jika kapasitas valid

warehouse = Warehouse(kapasitas)

# Menu interaktif
while True:
    print("\nMenu:")
    print("1. Tambah Barang")
    print("2. Ambil Barang")
    print("3. Lihat Barang di Gudang")
    print("4. Keluar")
    pilihan = input("Pilih menu (1/2/3/4): ")

    if pilihan == '1':
        barang = input("Masukkan nama barang: ")
        warehouse.add_item(barang)
    elif pilihan == '2':
        warehouse.remove_item()
    elif pilihan == '3':
        warehouse.show_items()
    elif pilihan == '4':
        print("Keluar dari program.")
        break
    else:
        print("Pilihan tidak valid. Silakan pilih lagi.")
```
