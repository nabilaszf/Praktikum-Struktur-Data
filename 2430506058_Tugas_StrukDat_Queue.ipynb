```python
# Implementasi Queue menggunakan List
class QueueList:
    def __init__(self):
        self.queue = []

    def enqueue(self, item):
        self.queue.append(item)

    def dequeue(self):
        if not self.is_empty():
            return self.queue.pop(0)
        return None

    def front(self):
        return self.queue[0] if not self.is_empty() else None

    def rear(self):
        return self.queue[-1] if not self.is_empty() else None

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

# Contoh penggunaan
queue = QueueList()
queue.enqueue(10)
queue.enqueue(20)
queue.enqueue(30)
print("Queue setelah enqueue:", queue.queue)
print("Dequeued element:", queue.dequeue())
print("Front element:", queue.front())
print("Rear element:", queue.rear())
print("Queue size:", queue.size())
```

```python
from collections import deque
# Implementasi Queue menggunakan deque
class QueueDeque:
    def __init__(self):
        self.queue = deque()

    def enqueue(self, item):
        self.queue.append(item)

    def dequeue(self):
        return self.queue.popleft() if not self.is_empty() else None

    def front(self):
        return self.queue[0] if not self.is_empty() else None

    def rear(self):
        return self.queue[-1] if not self.is_empty() else None

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

# Contoh penggunaan
queue = QueueDeque()
queue.enqueue(10)
queue.enqueue(20)
queue.enqueue(30)
print("Queue setelah enqueue:", list(queue.queue))
print("Dequeued element:", queue.dequeue())
print("Front element:", queue.front())
print("Rear element:", queue.rear())
print("Queue size:", queue.size())
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class QueueLinkedList:
    def __init__(self):
        self.front = self.rear = None

    def enqueue(self, data):
        new_node = Node(data)
        if self.rear is None:
            self.front = self.rear = new_node
            return
        self.rear.next = new_node
        self.rear = new_node

    def dequeue(self):
        if self.front is None:
            return None
        dequeued_element = self.front.data
        self.front = self.front.next
        if self.front is None:
            self.rear = None
        return dequeued_element

    def is_empty(self):
        return self.front is None

    def front_element(self):
        return self.front.data if self.front else None

    def rear_element(self):
        return self.rear.data if self.rear else None

# Contoh penggunaan
queue = QueueLinkedList()
queue.enqueue(10)
queue.enqueue(20)
queue.enqueue(30)
print("Dequeued element:", queue.dequeue())
print("Front element:", queue.front_element())
print("Rear element:", queue.rear_element())
```

```python
# LATIHAN 2

class AntrianFotokopi:
    def __init__(self):
        self.antrian = []

    def enqueue(self, nama, npm, num_pages):
        order_id = nama[:3].upper() + npm[-2:]
        if num_pages < 10:  # Fitur Fast Track
            print(f"Pesanan {order_id} ({num_pages} halaman) langsung diproses (Fast Track).")
        else:
            if num_pages > 100:  # Pesanan lebih dari 100 halaman diproses dalam dua tahap
                self.antrian.append({"id": order_id, "halaman": num_pages, "tahap": 1})
                print(f"Pesanan {order_id} ({num_pages} halaman) ditambahkan ke antrian (diproses dalam dua tahap).")
            else:
                self.antrian.append({"id": order_id, "halaman": num_pages, "tahap": 0})
                print(f"Pesanan {order_id} ({num_pages} halaman) ditambahkan ke antrian.")

    def dequeue(self):
        if not self.is_empty():
            pesanan = self.antrian[0]
            if pesanan["halaman"] > 100 and pesanan["tahap"] == 1:
                print(f"Melayani pesanan {pesanan['id']} (Tahap 1: 100 halaman).")
                pesanan["halaman"] -= 100
                pesanan["tahap"] = 2
                self.antrian.append(pesanan)  # Kembalikan ke antrian untuk tahap kedua
            else:
                print(f"Melayani pesanan {pesanan['id']} ({pesanan['halaman']} halaman).")
            return self.antrian.pop(0)
        return None

    def front(self):
        return self.antrian[0]["id"] if not self.is_empty() else None

    def rear(self):
        return self.antrian[-1]["id"] if not self.is_empty() else None

    def is_empty(self):
        return len(self.antrian) == 0

    def size(self):
        return len(self.antrian)

    def lihat_antrian(self):
        if self.is_empty():
            print("Antrian kosong.")
        else:
            print("Daftar Antrian:")
            for i, pesanan in enumerate(self.antrian, 1):
                status = "Tahap 1" if pesanan["halaman"] > 100 and pesanan["tahap"] == 1 else "Tahap 2" if pesanan["tahap"] == 2 else "Normal"
                print(f"{i}. {pesanan['id']} ({pesanan['halaman']} halaman, {status})")

antrian = AntrianFotokopi()

while True:
    print("\nMenu:")
    print("1. Tambah Pesanan")
    print("2. Layani Pesanan")
    print("3. Tampilkan Pesanan Pertama dan Terakhir")
    print("4. Cek Jumlah Pesanan")
    print("5. Lihat Daftar Antrian")
    print("6. Keluar")
    pilihan = input("Pilih menu: ")

    if pilihan == "1":
        nama = input("Masukkan tiga huruf pertama nama: ")
        npm = input("Masukkan dua digit terakhir NPM: ")
        num_pages = int(input("Masukkan jumlah halaman: "))
        antrian.enqueue(nama, npm, num_pages)
    elif pilihan == "2":
        pesanan = antrian.dequeue()
        if not pesanan:
            print("Antrian kosong.")
    elif pilihan == "3":
        print("Pesanan pertama:", antrian.front())
        print("Pesanan terakhir:", antrian.rear())
    elif pilihan == "4":
        print("Jumlah pesanan dalam antrian:", antrian.size())
    elif pilihan == "5":
        antrian.lihat_antrian()
    elif pilihan == "6":
        print("Keluar dari program.")
        break
    else:
        print("Pilihan tidak valid.")
```

```python
# LATIHAN 3

class AntrianParkir:
    def __init__(self, kapasitas=15):
        self.kapasitas = kapasitas
        self.antrian = []
        self.antrian_prioritas = []

    def enqueue(self, kota_lahir, npm, is_ramah_lingkungan=False):
        vehicle_number = kota_lahir[:2].upper() + npm[-2:]
        if is_ramah_lingkungan:
            self.antrian_prioritas.append(vehicle_number)
            print(f"Kendaraan {vehicle_number} (Ramah Lingkungan) ditambahkan ke antrian prioritas.")
        else:
            if self.size() < self.kapasitas:
                self.antrian.append(vehicle_number)
                print(f"Kendaraan {vehicle_number} ditambahkan ke antrian parkir.")
            else:
                print(f"Parkiran penuh, kendaraan {vehicle_number} harus menunggu.")

    def dequeue(self):
        if not self.is_empty_prioritas():
            kendaraan = self.antrian_prioritas.pop(0)
            print(f"Kendaraan {kendaraan} (Ramah Lingkungan) keluar dari parkiran.")
        elif not self.is_empty():
            kendaraan = self.antrian.pop(0)
            print(f"Kendaraan {kendaraan} keluar dari parkiran.")
        else:
            print("Tidak ada kendaraan dalam antrian.")

    def front(self):
        if not self.is_empty_prioritas():
            return self.antrian_prioritas[0]
        elif not self.is_empty():
            return self.antrian[0]
        return None

    def rear(self):
        if not self.is_empty():
            return self.antrian[-1]
        return None

    def is_empty(self):
        return len(self.antrian) == 0

    def is_empty_prioritas(self):
        return len(self.antrian_prioritas) == 0

    def size(self):
        return len(self.antrian) + len(self.antrian_prioritas)

    def lihat_antrian(self):
        if self.is_empty() and self.is_empty_prioritas():
            print("Antrian parkir kosong.")
        else:
            print("Daftar Antrian Parkir:")
            if not self.is_empty_prioritas():
                print("Prioritas (Ramah Lingkungan):")
                for i, kendaraan in enumerate(self.antrian_prioritas, 1):
                    print(f"{i}. {kendaraan}")
            if not self.is_empty():
                print("Reguler:")
                for i, kendaraan in enumerate(self.antrian, 1):
                    print(f"{i + len(self.antrian_prioritas)}. {kendaraan}")

parkiran = AntrianParkir()

while True:
    print("\nMenu:")
    print("1. Tambah Kendaraan")
    print("2. Keluarkan Kendaraan")
    print("3. Tampilkan Kendaraan Pertama dan Terakhir")
    print("4. Cek Jumlah Kendaraan")
    print("5. Lihat Daftar Antrian")
    print("6. Keluar")
    pilihan = input("Pilih menu: ")

    if pilihan == "1":
        kota_lahir = input("Masukkan dua huruf pertama kota lahir: ")
        npm = input("Masukkan dua digit terakhir NPM: ")
        ramah_lingkungan = input("Apakah kendaraan ramah lingkungan? (y/n): ").lower() == 'y'
        parkiran.enqueue(kota_lahir, npm, ramah_lingkungan)
    elif pilihan == "2":
        parkiran.dequeue()
    elif pilihan == "3":
        print("Kendaraan pertama:", parkiran.front())
        print("Kendaraan terakhir:", parkiran.rear())
    elif pilihan == "4":
        print("Jumlah kendaraan dalam antrian:", parkiran.size())
    elif pilihan == "5":
        parkiran.lihat_antrian()
    elif pilihan == "6":
        print("Keluar dari program.")
        break
    else:
        print("Pilihan tidak valid.")
```

```python
# LATIHAN 4

class AntrianSertifikat:
    def __init__(self):
        self.antrian_reguler = []
        self.antrian_berprestasi = []

    def enqueue(self, program_studi, npm, pernah_kompetisi_nasional=False):
        student_id = program_studi[0].upper() + npm[-2:]
        if pernah_kompetisi_nasional:
            print(f"Mahasiswa {student_id} (Berprestasi) langsung mengambil sertifikat.")
        else:
            if self.size() >= 20:
                self.antrian_berprestasi.append(student_id)
                print(f"Mahasiswa {student_id} ditambahkan ke jalur berprestasi.")
            else:
                self.antrian_reguler.append(student_id)
                print(f"Mahasiswa {student_id} ditambahkan ke jalur reguler.")

    def dequeue(self):
        if self.antrian_reguler:
            mahasiswa = self.antrian_reguler.pop(0)
            print(f"Mahasiswa {mahasiswa} (Reguler) telah dilayani.")
        elif self.antrian_berprestasi:
            mahasiswa = self.antrian_berprestasi.pop(0)
            print(f"Mahasiswa {mahasiswa} (Berprestasi) telah dilayani.")
        else:
            print("Tidak ada mahasiswa dalam antrian.")

    def front(self):
        if self.antrian_reguler:
            return self.antrian_reguler[0]
        elif self.antrian_berprestasi:
            return self.antrian_berprestasi[0]
        return None

    def rear(self):
        if self.antrian_berprestasi:
            return self.antrian_berprestasi[-1]
        elif self.antrian_reguler:
            return self.antrian_reguler[-1]
        return None

    def is_empty(self):
        return len(self.antrian_reguler) == 0 and len(self.antrian_berprestasi) == 0

    def size(self):
        return len(self.antrian_reguler) + len(self.antrian_berprestasi)

    def lihat_antrian(self):
        if self.is_empty():
            print("Antrian kosong.")
        else:
            print("Daftar Antrian:")
            if self.antrian_reguler:
                print("Jalur Reguler:")
                for i, mahasiswa in enumerate(self.antrian_reguler, 1):
                    print(f"{i}. {mahasiswa}")
            if self.antrian_berprestasi:
                print("Jalur Berprestasi:")
                for i, mahasiswa in enumerate(self.antrian_berprestasi, 1):
                    print(f"{i + len(self.antrian_reguler)}. {mahasiswa}")

antrian = AntrianSertifikat()

while True:
    print("\nMenu:")
    print("1. Tambah Mahasiswa")
    print("2. Layani Mahasiswa")
    print("3. Tampilkan Mahasiswa Pertama dan Terakhir")
    print("4. Cek Jumlah Mahasiswa")
    print("5. Lihat Daftar Antrian")
    print("6. Keluar")
    pilihan = input("Pilih menu: ")

    if pilihan == "1":
        program_studi = input("Masukkan program studi: ")
        npm = input("Masukkan dua digit terakhir NPM: ")
        kompetisi = input("Apakah pernah mengikuti kompetisi nasional? (y/n): ").lower() == 'y'
        antrian.enqueue(program_studi, npm, kompetisi)
    elif pilihan == "2":
        antrian.dequeue()
    elif pilihan == "3":
        print("Mahasiswa pertama:", antrian.front())
        print("Mahasiswa terakhir:", antrian.rear())
    elif pilihan == "4":
        print("Jumlah mahasiswa dalam antrian:", antrian.size())
    elif pilihan == "5":
        antrian.lihat_antrian()
    elif pilihan == "6":
        print("Keluar dari program.")
        break
    else:
        print("Pilihan tidak valid.")
```
